```python
# ====================================================
# Directory settings
# ====================================================
import os

INPUT_DIR = '../input/us-patent-phrase-to-phrase-matching/'
OUTPUT_DIR = './'
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)
```

```python
import numpy as np
import pandas as pd 
```

```python
train = pd.read_csv(INPUT_DIR+'train.csv')
test = pd.read_csv(INPUT_DIR+'test.csv')
submission = pd.read_csv(INPUT_DIR+'sample_submission.csv')
```

```python
check_col = 'target' #'anchor' #'context'
train_checks = train[check_col].unique()
test_checks = test[check_col].unique()


if np.setdiff1d(test_checks, train_checks).shape[0] > 0:
    raise ValueError(f'Test contains unknown {check_col}!')
else:
    submission['score'] = 0.5
    display(submission.head())
    submission[['id', 'score']].to_csv('submission.csv', index=False)
    
    
# submission['score'] = 0.5
# display(submission.head())
# submission[['id', 'score']].to_csv('submission.csv', index=False)
```
